## 2.Application n°1 : Création de recettes 🍳
### 2.1.Consigne du système (ne change pas)
C'est la "consigne" (ou le "prompt" en anglais) que l'on va envoyer à ChatGPT avec une saisie de l'utilisateur. C'est une très bonne manière de rendre les réponses de ChatGPT plus précises, mais aussi plus reproducibles.

```python

fixed_prompt =  """
Tu vas générer une recette complète incluant un petit texte d'introduction, une liste d'ingrédients, les quantités pour ces ingrédients et les instructions pour la recette suivante :
"""
```

### 2.2.Saisie de l'utilisateur (à vous d'être créatif)
Entrez le nom de votre recette dans la boîte à droite ▶


```python
User_prompt = 'des canelés sans rhum' # @param {type:"string"}
```

```python
#@title 2.3.Envoyons la requête à ChatGPT

completion = client.chat.completions.create(
  model = model_to_use,
  messages=[
    {"role": "system", "content": fixed_prompt},
   {"role": "user", "content": User_prompt}
  ]
)

openai_response = completion.choices[0].message.content

print(openai_response)
```

### 2.4.Mais combien ça coûte ?

Les "tokens" sont l'unité de mesure utilisée par OpenAI pour mesurer la taille de la requête. Une bonne règle générale est que 10 mots équivalent typiquement à 7 tokens.

```python
print(f"Nombre de tokens utilisés : {completion.usage.total_tokens}")

price_per_token = 0.5/1000000
querry_cost = completion.usage.total_tokens * price_per_token

print(f"Cette requête a coûté : {querry_cost} $")
```

```python
#@title 3.1.3.Nous allons faire une liste des noms de toutes les bactéries trouvées dans notre analyse du microbiome


# C'est ici que nous faisons la liste (elle s'appelle "bacteria_list". Très original, je sais...)
bacteria_list = df.columns

# Pour gagner du temps, raccourcissons notre liste, vous pouvez décider de la longueur de la liste avec le paramètre ci-dessous :
number_of_bacteria_to_keep = 8
bacteria_list = bacteria_list[:number_of_bacteria_to_keep]

print(bacteria_list)
```

Nous avons bien une liste mais tous les noms sont étrangement formatés. Faisons un petit nettoyage ! 🧹

## 3.2.Application n°2 : Descriptions automatisées de noms de bactéries 🦠
### 3.2.1.Consigne du système (ne change pas)
C'est la "consigne" (ou le "prompt" en anglais) que l'on va envoyer à ChatGPT avec une saisie de l'utilisateur. C'est une très bonne manière de rendre les réponses de ChatGPT plus précises, mais aussi plus reproducibles.

```python
fixed_prompt =  """
Imprime les informations les plus scientifiquement précises pour les bactéries suivantes. En particulier, inclue des informations pertinentes sur le rôle des bactéries suivantes dans la santé humaine et la régulation du microbiome.
"""
```

```python
#@title 3.2.2.Paramètres pour la requête ChatGPT
model_to_use = "gpt-3.5-turbo"
# model_to_use = "gpt-4"

# Selon à quel point vous voulez que la réponse soit aléatoire, vous pouvez modifier la température
# (min=0 est très déterministique, max=2 est très aléatoire, nous choisirons 0.2 mais n'hésitez pas à essayer autre chose)
temperature = 0.2
```

```python
#@title 3.2.3.Envoyons la requête à ChatGPT

output = list()
full_output= list()

# La partie importante est la ligne suivante
# Nous l'appelons une "boucle for", elle nous permet d'itérer la requête pour chaque bactérie de notre liste à l'API
for bac in bacteria_list:
  print(f'\n------------------------ {bac} ------------------------\n')

  bacteria_short = bac.split(";")[-1]
  if bacteria_short != "__":
    bacteria_short = bacteria_short.split("__")[-1]
    User_prompt = bacteria_short
    message_content = []

    message_content.append({"role": "system", "content": fixed_prompt})

    message_content.append({"role": "user", "content": User_prompt})

    completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
      {"role": "user", "content": fixed_prompt + " " + User_prompt}
      ]
    )

    openai_response = completion.choices[0].message.content
    output.append(openai_response)
    full_output.append(completion)

    for line in openai_response.split(". "):
      print(line)

    # print(openai_response)

  else:
    print("Missing taxonomic information.")
```

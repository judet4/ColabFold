## 2.Application n¬∞1 : Cr√©ation de recettes üç≥
### 2.1.Consigne du syst√®me (ne change pas)
C'est la "consigne" (ou le "prompt" en anglais) que l'on va envoyer √† ChatGPT avec une saisie de l'utilisateur. C'est une tr√®s bonne mani√®re de rendre les r√©ponses de ChatGPT plus pr√©cises, mais aussi plus reproducibles.

```python

fixed_prompt =  """
Tu vas g√©n√©rer une recette compl√®te incluant un petit texte d'introduction, une liste d'ingr√©dients, les quantit√©s pour ces ingr√©dients et les instructions pour la recette suivante :
"""
```

### 2.2.Saisie de l'utilisateur (√† vous d'√™tre cr√©atif)

```python
User_prompt = 'des canel√©s sans rhum' # @param {type:"string"}
```

```python
#@title 2.3.Envoyons la requ√™te √† ChatGPT

completion = client.chat.completions.create(
  model = model_to_use,
  messages=[
    {"role": "system", "content": fixed_prompt},
   {"role": "user", "content": User_prompt}
  ]
)

openai_response = completion.choices[0].message.content

print(openai_response)
```

### 2.4.Mais combien √ßa co√ªte ?

Les "tokens" sont l'unit√© de mesure utilis√©e par OpenAI pour mesurer la taille de la requ√™te. Une bonne r√®gle g√©n√©rale est que 10 mots √©quivalent typiquement √† 7 tokens.

```python
print(f"Nombre de tokens utilis√©s : {completion.usage.total_tokens}")

price_per_token = 0.5/1000000
querry_cost = completion.usage.total_tokens * price_per_token

print(f"Cette requ√™te a co√ªt√© : {querry_cost} $")
```

## 3.2.Application n¬∞2 : Descriptions automatis√©es de noms de bact√©ries ü¶†

```python
fixed_prompt =  """
Imprime les informations les plus scientifiquement pr√©cises pour les bact√©ries suivantes. En particulier, inclue des informations pertinentes sur le r√¥le des bact√©ries suivantes dans la sant√© humaine et la r√©gulation du microbiome.
"""
```

```python
#@title 3.2.2.Param√®tres pour la requ√™te ChatGPT
model_to_use = "gpt-3.5-turbo"
# model_to_use = "gpt-4"

# Selon √† quel point vous voulez que la r√©ponse soit al√©atoire, vous pouvez modifier la temp√©rature
# (min=0 est tr√®s d√©terministique, max=2 est tr√®s al√©atoire, nous choisirons 0.2 mais n'h√©sitez pas √† essayer autre chose)
temperature = 0.2
```

```python
#@title 3.2.3.Envoyons la requ√™te √† ChatGPT

output = list()
full_output= list()

# La partie importante est la ligne suivante
# Nous l'appelons une "boucle for", elle nous permet d'it√©rer la requ√™te pour chaque bact√©rie de notre liste √† l'API
for bac in bacteria_list:
  print(f'\n------------------------ {bac} ------------------------\n')

  bacteria_short = bac.split(";")[-1]
  if bacteria_short != "__":
    bacteria_short = bacteria_short.split("__")[-1]
    User_prompt = bacteria_short
    message_content = []

    message_content.append({"role": "system", "content": fixed_prompt})

    message_content.append({"role": "user", "content": User_prompt})

    completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
      {"role": "user", "content": fixed_prompt + " " + User_prompt}
      ]
    )

    openai_response = completion.choices[0].message.content
    output.append(openai_response)
    full_output.append(completion)

    for line in openai_response.split(". "):
      print(line)

    # print(openai_response)

  else:
    print("Missing taxonomic information.")
```
